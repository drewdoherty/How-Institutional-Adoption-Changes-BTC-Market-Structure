```python
from pathlib import Path
import pandas as pd

candidates = [Path.cwd(), *Path.cwd().parents]
root = None
for cand in candidates:
    if (cand / 'reports' / 'transparency_shift').exists():
        root = cand
        break
if root is None:
    raise RuntimeError('Could not locate project root with reports/transparency_shift')
reports_root = root / 'reports' / 'transparency_shift'
run_dirs = sorted([p for p in reports_root.glob('*') if p.is_dir()])
if not run_dirs:
    raise RuntimeError('No runs found. Run scripts/analyze_transparency_shift.py first.')
latest = run_dirs[-1]
latest
```
